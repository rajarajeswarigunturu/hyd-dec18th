```python
# if string first letter is s print same
# otherwise print z in another seperate column
```

```python
def test1(a):
    if a=='s':
        return 's'
    else:
        return 'z'
```

```python
df['col6']=df['col5'].apply(test1)
```

```python
df
```

# by using lambda function

```python
df['col6_a']=df['col5'].apply(lambda a: 's' if a=='s' else 'z')
```
